```python
import pandas as pd
```

```python
# Import both CSVs. Convert all date columns to datetime. 
# Check for missing values in the visits data. 
# Fill missing Insurance_Covered with the median Insurance_Covered for that specific Department using .transform(). 
# Fill missing Wait_Time_Min with the average Wait_Time_Min for that specific Doctor_ID using .transform(). 
# Verify no nulls remain.

patients = pd.read_csv('hospital_visits.csv')
patients['Visit_Date'] = pd.to_datetime(patients['Visit_Date'])
patients.isnull().sum()
patients['Insurance_Covered'] = patients['Insurance_Covered'].fillna(patients.groupby('Department')['Insurance_Covered'].transform('median'))
patients['Wait_Time_Min'] = patients['Wait_Time_Min'].fillna(patients.groupby('Doctor_ID')['Wait_Time_Min'].transform('mean'))
patients.isnull().sum()
patients
```

```python
doctors = pd.read_csv('doctors.csv')
doctors['Hire_Date'] = pd.to_datetime(doctors['Hire_Date'])
doctors
```

```python
# Inner join the visits with the doctors data on Doctor_ID. 
# Then do a left join on the same data. 
# Compare the row counts — identify which transactions exist in the left join but not the inner join and explain why. 
# Create a column Doctor_Matched using .notna().

patdocinner = patients.merge(doctors, on = 'Doctor_ID', how = 'left')
patdocinner.shape[0]
patdocinner.shape[1]
patdocinner['Doctor_Matched'] = patdocinner['Doctor_Name'].notna()
patdocinner
```

```python
patdocleft = patients.merge(doctors, on = 'Doctor_ID', how = 'left')
patdocleft.shape[0]
patdocleft.shape[1]
patdocleft['Doctor_Matched'] = patdocleft['Doctor_Name'].notna()
patdocleft
```

```python
# Create columns for Out_Of_Pocket (Treatment_Cost - Insurance_Covered), and Cost_Ratio (Insurance_Covered / Treatment_Cost). 
# Convert Follow_Up_Required to boolean. 
# Then create a pivot table showing the average Out_Of_Pocket by Department (rows) and Follow_Up_Required (columns) with margins. 
# For each Department, calculate the cumulative Treatment_Cost ordered by Visit_Date. Also calculate the visit-over-visit percentage change in Treatment_Cost per Department using .shift().

patdoc = patdocleft.copy()
patdoc['Out_Of_Pocket'] = patdoc['Treatment_Cost'] - patdoc['Insurance_Covered']
patdoc['Cost_Ratio'] = (patdoc['Insurance_Covered'] / patdoc['Treatment_Cost']).round(2)
pt = pd.pivot_table(patdoc, values = 'Out_Of_Pocket', index = 'Department', columns = 'Follow_Up_Required', aggfunc = 'mean', fill_value = 0, margins = True).round(2)
pt

patdoc = patdoc.sort_values('Visit_Date')
patdoc['Cumulative_Treatment_Cost'] = patdoc.groupby('Department')['Treatment_Cost'].cumsum()
patdoc['Visit_Percent_Change'] = (patdoc['Treatment_Cost'] / (patdoc.groupby('Department')['Treatment_Cost'].shift(1)) * 100).round(2)
patdoc
```

```python
# Using .agg() with named aggregations, calculate per Doctor: total Treatment_Cost, average Wait_Time_Min, number of unique patients (nunique), and follow-up rate (mean of the boolean Follow_Up_Required). 
# Rank doctors by total Treatment_Cost descending. Use .map() with a dictionary to label doctors with total Treatment_Cost > 10000 as 'High Volume' and others as 'Standard'.

patdoc['Follow_Up_Required'] = patdoc['Follow_Up_Required'].apply(lambda x: True if x == 'Yes' else False)
doctor_stats = patdoc.groupby('Doctor_ID').agg(Total_Treatment_Cost = ('Treatment_Cost', 'sum'), Avergae_Wait = ('Wait_Time_Min', 'mean'), Unique_Patients = ('Patient_ID', 'nunique'), Follow_Up_Rate = ('Follow_Up_Required', 'mean'))
doctor_stats['Rank'] = doctor_stats['Total_Treatment_Cost'].rank(ascending = False).astype(int)
performance_dict = {doctor: 'High Volume' if cost > 10000 else 'Standard' for doctor, cost in doctor_stats['Total_Treatment_Cost'].items()}
doctor_stats['Performance'] = doctor_stats.index.map(performance_dict)
doctor_stats
```

```python
# Identify repeat patients (more than one visit) using boolean indexing on grouped results. 
# For repeat patients only, find each one's most common Diagnosis using .drop_duplicates() after sorting. 
# Calculate their average Out_Of_Pocket and total visits. 
# Find which repeat patient has the highest cumulative Treatment_Cost using .idxmax(). 
# Display everything using a nested lambda to label patients: 'Frequent' if visits > 3, 'Returning' if 2-3 visits.

patient_stats = patdoc.groupby('Patient_ID').size().reset_index(name = 'Number_of_Visits')
patient_stats = patient_stats[patient_stats['Number_of_Visits'] > 1]

diagnosis_count = patdoc.groupby(['Patient_ID', 'Diagnosis']).size().reset_index(name = 'Diagnosis_Count')
top_diagnosis = diagnosis_count.sort_values('Diagnosis_Count', ascending=False)
top_diagnosis = top_diagnosis.drop_duplicates(subset='Patient_ID')
top_diagnosis

repeat_oop = (patdoc[patdoc['Patient_ID'].isin(patient_stats['Patient_ID'])].groupby('Patient_ID')['Out_Of_Pocket'].mean().reset_index(name='Avg_Out_Of_Pocket')).round(2)

final = patient_stats.merge(top_diagnosis[['Patient_ID', 'Diagnosis']], on='Patient_ID')
final = final.merge(repeat_oop, on='Patient_ID')

pctc = patdoc[patdoc['Patient_ID'].isin(patient_stats['Patient_ID'])].groupby('Patient_ID')['Treatment_Cost'].sum().reset_index(name='Total_Treatment_Cost')
final = final.merge(pctc, on = 'Patient_ID')
final = final.sort_values('Total_Treatment_Cost', ascending = False)
expensive = final.loc[final['Total_Treatment_Cost'].idxmax(), 'Patient_ID']
print(f"The patient with the highest cumulative treatment costs is: {expensive}")

final['Patient_Visit_Rating'] = final['Number_of_Visits'].apply(lambda x: 'Frequent' if x > 3 else ('Returning' if x >= 2 else ('Not Frequent' if x == 1 else 'Non-existent')))
final
```
