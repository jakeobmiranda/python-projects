```python
import pandas as pd
```

```python
# Import the CSV and convert Transaction_Date to a proper datetime type.
df = pd.read_csv('transactions.csv')
df['Transaction_Date'] = pd.to_datetime(df['Transaction_Date'])

# Create a column called Transaction_Type — label it 'Purchase' if Amount is positive, 'Return' if negative. Use a lambda.
df['Transaction_Type'] = df['Amount'].apply(lambda x: 'Purchase' if x > 0 else 'Return')
df
```

```python
# For each Customer_ID, calculate their total spend (sum of all positive amounts only), total returns (sum of all negative amounts only), 
# and net spend (total of everything). Display this as a single grouped DataFrame sorted by net spend descending.
total_spend = df[df['Amount'] > 0].groupby('Customer_ID')['Amount'].sum()
total_returns = df[df['Amount'] < 0].groupby('Customer_ID')['Amount'].sum()
net_spend = df.groupby('Customer_ID')['Amount'].sum()

result = pd.DataFrame({'Total_Spend' : total_spend, 'Total_Returns' : total_returns, 'Net_Spend' : net_spend})
result.sort_values('Net_Spend', ascending=False)
```

```python
# Create a column called Cumulative_Spend that shows the running total of Amount for each customer, ordered by Transaction_Date. 
# This means for each row, the value should be the sum of that customer's amounts from their first transaction up to and including that row.

df['Cumulative_Spend'] = df.groupby('Customer_ID')['Amount'].cumsum()
df
```

```python
# Identify which customer has the highest single transaction amount and which customer has the most transactions overall. 
# Print both answers with labels.

maxamount = df.loc[df['Amount'].idxmax(), 'Customer_ID']
mosttransactions = df.groupby('Customer_ID')['Transaction_ID'].count().idxmax()

print(f"The most expensive transaction was from Customer ID: {maxamount} and the Customer ID with the most transactions was: {mosttransactions}.")
```
