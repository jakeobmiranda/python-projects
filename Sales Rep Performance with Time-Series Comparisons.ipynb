```python
import pandas as pd
```

```python
# Import the CSV, convert Start_Date to datetime, and create a column called Tenure_Days that calculates how many days each rep has been with the company as of 2024-12-31.
df = pd.read_csv('sales_reps.csv')
df['Start_Date'] = pd.to_datetime(df['Start_Date'])
df['Tenure_Days'] = (pd.to_datetime('2024-12-31') - df['Start_Date']).dt.days
df
```

```python
# Create a column called Win_Rate calculated as Deals_Closed / (Deals_Closed + Deals_Lost). Then create a Win_Rate_Label using a lambda: 'Strong' if Win_Rate > 0.7, 'Average' if between 0.5 and 0.7 inclusive, 'Weak' if below 0.5.
df['Win_Rate'] = (df['Deals_Closed'] / (df['Deals_Closed'] + df['Deals_Lost'])).round(2)
df['Win_Rate_Label'] = df['Win_Rate'].apply(lambda x: 'Strong' if x > 0.7 else ('Weak' if x < 0.5 else 'Average'))
df
```

```python
# Create a column called Revenue_Change that shows the quarter-over-quarter percentage change in Revenue for each rep. Q1 should show NaN since there's no prior quarter. The formula is ((Current - Previous) / Previous) * 100.
df['Revenue_Change'] = ((df['Revenue'] - df.groupby('Rep_ID')['Revenue'].shift(1)) / (df.groupby('Rep_ID')['Revenue'].shift(1)) * 100).round(2)
df
```

```python
# For each Manager, calculate the total Revenue, average Win_Rate, and total Deals_Closed across all their reps. Display sorted by total Revenue descending.
#x = df.groupby('Manager')['Revenue'].sum()
#y = df.groupby('Manager')['Win_Rate'].mean()
#z = df.groupby('Manager')['Deals_Closed'].sum()

#result = pd.DataFrame({'Manager_Total_Rev' : x, 'Manager_Win_Avg' : y, 'Manager_Deals_Closed' : z})
#result = result.sort_values('Manager_Total_Rev', ascending=False)

#result = x.merge(y, on='Manager').merge(z, on='Manager')
#result.sort_values('Manager_Total_Rev', ascending=False)

result = df.groupby('Manager').agg(Manager_Total_Rev=('Revenue', 'sum'), Manager_Win_Avg=('Win_Rate', 'mean'), Manager_Deals_Closed=('Deals_Closed', 'sum')).reset_index().sort_values('Manager_Total_Rev', ascending=False)
result
```

```python
# Rank each rep by their total annual Revenue (sum across all quarters) and display the ranking with rep name and total revenue. Rank 1 = highest revenue.
rep_rank = df.groupby('Rep_Name')['Revenue'].sum().reset_index(name='Total_Revenue')
rep_rank['Rank'] = rep_rank['Total_Revenue'].rank(ascending=False).astype(int)
rep_rank = rep_rank.sort_values('Rank')
rep_rank
```
